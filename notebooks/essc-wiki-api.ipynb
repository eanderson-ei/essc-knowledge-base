```python
# read newline json pattern file for tags
tags = []

with open(r'C:/Users/Erik/Documents/_dev/essc-knowledge-base/data/inputs/entity-patterns.jsonl', 'r') as f:
    for line in f:
        entry = json.loads(line)
        combined = []
        for lower in entry.get('pattern'):
            combined.extend(list(lower.values()))
        tags.append(' '.join(combined))
        tags.append(entry.get('id').lower())
tags = list(set(tags))
tags
```
